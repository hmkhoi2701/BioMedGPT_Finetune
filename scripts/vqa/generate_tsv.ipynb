```python
import os
import pandas as pd
import json
```

```python
mimic_path = '/home/khoihm/khoihm/data/mimic_zip'
vindr_path = '/home/khoihm/khoihm/data/VinDr/vindr_zip_new'
mimic_train_path = os.path.join(mimic_path,'annotation_resampled_new_3/train')
mimic_test_path = os.path.join(mimic_path,'annotation_resampled_new_3/test')
vindr_train_path = os.path.join(vindr_path,'annotation_resampled_new_2/train')
vindr_test_path = os.path.join(vindr_path,'annotation_resampled_new_2/test')
train_df = {'id':[],'sid':[],'question':[],'answer':[],'class':[],'img_path':[]}
test_df = {'id':[],'sid':[],'question':[],'answer':[],'class':[],'img_path':[]}
```

```python
train_id = 0
for file in os.listdir(mimic_train_path):
    if file.endswith('.json'):
        with open(os.path.join(mimic_train_path,file)) as f:
            data = json.load(f)
            train_df['id'].append(train_id)
            train_df['sid'].append(data['shapes'][0]['image_name'])
            train_df['question'].append(data['text'].lower())
            train_df['answer'].append('1|!+'+data['answer'].lower()[data['answer'].lower().find('seg')+6:])
            train_df['class'].append('cxr')
            train_df['img_path'].append(os.path.join(mimic_path,'files',data['shapes'][0]['image_name']))
    train_id += 1
for file in os.listdir(vindr_train_path):
    if file.endswith('.json'):
        with open(os.path.join(vindr_train_path,file)) as f:
            data = json.load(f)
            train_df['id'].append(train_id)
            train_df['sid'].append(data['shapes'][0]['image_name'])
            train_df['question'].append(data['text'].lower())
            train_df['answer'].append('1|!+'+data['answer'].lower()[data['answer'].lower().find('seg')+6:])
            train_df['class'].append('cxr')
            train_df['img_path'].append(os.path.join(vindr_path,'jpg/all',data['shapes'][0]['image_name']))
    train_id += 1

test_id = 0
for file in os.listdir(mimic_test_path):
    if file.endswith('.json'):
        with open(os.path.join(mimic_test_path,file)) as f:
            data = json.load(f)
            test_df['id'].append(test_id)
            test_df['sid'].append(data['shapes'][0]['image_name'])
            test_df['question'].append(data['text'].lower())
            test_df['answer'].append('1|!+'+data['answer'].lower()[data['answer'].lower().find('seg')+6:])
            test_df['class'].append('cxr')
            test_df['img_path'].append(os.path.join(mimic_path,'files',data['shapes'][0]['image_name']))
    test_id += 1
for file in os.listdir(vindr_test_path):
    if file.endswith('.json'):
        with open(os.path.join(vindr_test_path,file)) as f:
            data = json.load(f)
            test_df['id'].append(test_id)
            test_df['sid'].append(data['shapes'][0]['image_name'])
            test_df['question'].append(data['text'].lower())
            test_df['answer'].append('1|!+'+data['answer'].lower()[data['answer'].lower().find('seg')+6:])
            test_df['class'].append('cxr')
            test_df['img_path'].append(os.path.join(vindr_path,'jpg/all',data['shapes'][0]['image_name']))
    test_id += 1

```

```python
test_df = pd.DataFrame(test_df)
train_df = pd.DataFrame(train_df)
```
